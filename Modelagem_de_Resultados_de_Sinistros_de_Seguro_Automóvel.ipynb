## Modelagem de Resultados de Sinistros de Seguro Automóvel

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit
```

### Leitura e apresentação do conjunto de dados

```python
cars = pd.read_csv("car_insurance.csv")
cars
```

### Verificando informações sobre o dataframe cars

```python
cars.info()
```

O conjunto de dados possui 18 colunas e 10000 índices, ou seja, 10000 valores por coluna. Tendo isso em mente podemos perceber que as colunas *credit_score* e *annual_mileage* possui valores nulos, nesse caso precisamos fazer um tratamento adequado para o conjunto de dados. Vamos substituí os valores nulos pela média dos valores de cada coluna

### Preenchimento de valores ausentes com a média

```python
cars["credit_score"].fillna(cars["credit_score"].mean(), inplace=True)
cars["annual_mileage"].fillna(cars["annual_mileage"].mean(), inplace=True)
```

Esta técnica ajuda a manter a distribuição geral dos dados

```python
cars.info()
```

### Criação de modelos de regressão logística

tendo em vista que a coluna "id" é um identificador único e "outcome" é a variável de resposta que está sendo prevista, vamos excluí-las da atribuição à variavel *features*

```python
models = []
features = cars.drop(columns=["id", "outcome"]).columns

for col in features:
    model = logit(f"outcome ~ {col}", data=cars).fit()
    models.append(model)
```

criado modelos de regressão logística, ajustados com o *fit()* e adicionado à variavel *models*

### Calcular as precisões (accuracies) para cada modelo

```python
accuracies = []

for feature in range(0, len(models)):
    conf_matrix = models[feature].pred_table()
    tn = conf_matrix[0,0]
    tp = conf_matrix[1,1]
    fn = conf_matrix[1,0]
    fp = conf_matrix[0,1]
    acc = (tn + tp) / (tn + fn + fp + tp)
    accuracies.append(acc)
```

### Selecionar a característica com a maior precisão

```python
best_feature = features[accuracies.index(max(accuracies))]
```

### Criação de um DataFrame com a melhor característica e precisão

Armazenar como um DataFrame chamado *best_feature_df*, contendo colunas nomeadas *best_feature* e *best_accuracy* com o nome do recurso de maior precisão, e a respectiva pontuação de precisão.

```python
best_feature_df = pd.DataFrame({"best_feature": best_feature,
                                "best_accuracy": max(accuracies)},
                                index=[0])
best_feature_df
```
